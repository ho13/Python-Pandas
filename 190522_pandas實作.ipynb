```python
import pandas as pd
import numpy as np
from sas7bdat import SAS7BDAT

ex2 = SAS7BDAT("P:\\07_Python學習筆記\\Ye Ya講義\\190522\\dataset\\income_101yr.sas7bdat", encoding = "big5").to_data_frame()
ex2.head(3)
```

```python
income = ex2[['city_distinct','tax_unit','income']]
res1 = income.groupby(['city_distinct']).sum()
res1.head()
```

```python
res1 = res1.reset_index()
res1.head(3)
```

```python
area = SAS7BDAT("P:\\07_Python學習筆記\\Ye Ya講義\\190522\\dataset\\area.sas7bdat", encoding="big5").to_data_frame()
area.head(3)
```

```python
res = res1.merge(area, left_on = 'city_distinct', right_on = 'city_town', how = 'inner')
res.head(3)
```

```python
res['avg_tax'] = res['tax_unit'] / res['area(km2)']
res['avg_income'] = res['income'] / res['area(km2)']
res.head(3)
```

```python
res.sort_values(['avg_income'], ascending = [False]).head(3)
```
